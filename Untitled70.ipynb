```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df=pd.read_csv('/content/twitter_training.csv (1).zip')
df.info()
```

```python
df.head()
```

```python
# Inspect how many columns are there
print("Number of columns:", df.shape[1])
df.head()
```

```python
df.columns = ['id', 'entity', 'sentiment', 'content']
df.head()
```

```python
sns.countplot(x='sentiment', data=df, hue='sentiment', palette='viridis', legend=False)

plt.title("Sentiment Distribution")
plt.xlabel('sentiment')
plt.ylabel('count')
plt.show()
```

```python
# Count the frequency of each entity
top_entities = df['entity'].value_counts().head(10)

# Display as a bar chart
top_entities.plot(kind='bar', color='pink')
plt.title("Top 10 Most Mentioned Entities")
plt.xlabel("Entity")
plt.ylabel("Number of Mentions")
plt.tight_layout()
plt.show()
```

```python
# Clean the 'content' column by filling missing values with empty strings and converting to string type
df['content'] = df['content'].fillna('').astype(str)
```

```python
from wordcloud import WordCloud

positive_text = ' '.join(df[df['sentiment'] == 'Positive']['content'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud - Positive Tweets')
plt.show()
```

```python
from collections import Counter

positive_words = ' '.join(df[df['sentiment'] == 'Positive']['content']).split()
counter = Counter(positive_words)
print(counter.most_common(20))
```
