```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/India accident prediction.csv")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
# step 1
# Assuming each row represents one accident, we can count the number of accidents per state
state_accidents = df["State Name"].value_counts().sort_values(ascending=False)

# Select the top 10 states
top_10_states = state_accidents.head(10)

# Plotting the top 10 states
plt.figure(figsize=(12,6))
top_10_states.plot(kind='bar', color='tomato')
plt.title("Top 10 States by Total Accidents")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Step 2: Accident Severity Distribution
severity_counts = df['Accident Severity'].value_counts()

severity_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'gold', 'lightgreen'])
plt.title("Accident Severity Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
#Step 3: Monthly Accident Trend
monthly = df['Month'].value_counts().reindex([
    'January','February','March','April','May','June','July','August','September','October','November','December'])

monthly.plot(kind='line', marker='o', color='teal')
plt.title("Monthly Accident Trend")
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Step 4: Weather Condition Impact
weather_counts = df['Weather Conditions'].value_counts().head(10)

weather_counts.plot(kind='barh', color='orchid')
plt.title("Top 10 Weather Conditions During Accidents")
plt.xlabel("Number of Accidents")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#Step 5: Driver Gender Distribution
gender_counts = df['Driver Gender'].value_counts()

gender_counts.plot(kind='bar', color='mediumseagreen')
plt.title("Driver Gender Involved in Accidents")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#step 6
#License Status vs Accident Severity (Grouped Bar Chart)
license_severity = df[df['Driver License Status'].notnull()] \
    .groupby(['Driver License Status', 'Accident Severity']) \
    .size().unstack().fillna(0)

license_severity.plot(kind='bar', stacked=False, colormap='Set2')
plt.title("License Status vs Accident Severity")
plt.xlabel("Driver License Status")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#step 7
# Alcohol Involvement (Pie Chart)
alcohol_counts = df['Alcohol Involvement'].value_counts()

alcohol_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['salmon', 'lightblue'])
plt.title("Alcohol Involvement in Accidents")
plt.ylabel("")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# step 8
#Fatalities vs Casualties by State
state_fatal_casual = df.groupby("State Name")[["Number of Fatalities", "Number of Casualties"]].sum().sort_values(by="Number of Fatalities", ascending=False)

state_fatal_casual.plot(kind='bar', figsize=(12,6), stacked=True, colormap='Set2')
plt.title("Fatalities and Casualties by State")
plt.ylabel("Number of People")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
# step 9
# Heatmap - State vs Month
pivot_heat = df.pivot_table(index='State Name', columns='Month', values='Year', aggfunc='count') # Use count on any column to get the count

plt.figure(figsize=(12,8))
sns.heatmap(pivot_heat, cmap='YlOrRd', linewidths=0.5, linecolor='gray', annot=True, fmt='.0f')
plt.title("Heatmap: Monthly Accidents by State")
plt.xlabel("Month")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
# step 10
# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df[["Number of Casualties", "Number of Fatalities", "Number of Vehicles Involved"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation between Casualties, Fatalities, and Vehicles Involved")
plt.show()
```
